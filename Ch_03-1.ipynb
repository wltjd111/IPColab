```python
img = cv2.imread('/content/gdrive/My Drive/images/aero2.bmp')
print(img.shape)
gimg1 = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gimg1, cmap='gray')
```

```python
img = cv2.imread('/content/gdrive/My Drive/images/milkdrop.bmp')
print(img.shape)
gimg2 = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gimg2, cmap='gray')
```

```python
gPlus = np.zeros((256, 256), dtype=np.ubyte)
```

```python
for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    #gPlus[h,w] = gimg1[h,w] + gimg2[h,w]
    pValue = (np.int32(gimg1[h,w]) + np.int32(gimg2[h,w]))
    if pValue > 255:
      gPlus[h,w] = 255
    else:
      gPlus[h,w] = pValue

plt.imshow(gPlus, cmap='gray')
```

```python
for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    gPlus[h,w] = gimg1[h,w] * 0.5 + gimg2[h,w] * 0.5

plt.imshow(gPlus, cmap='gray')
```

```python
gPlus = cv2.add(gimg1, gimg2)
plt.imshow(gPlus, cmap='gray')
```

```python

gMinus = np.zeros((gimg1.shape[0], gimg1.shape[1]), dtype=np.ubyte)
```

```python
for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    gMinus[h,w] = np.abs(np.int32(gimg1[h,w]) - np.int32(gimg2[h,w]))

plt.imshow(gMinus, cmap='gray')
```

```python
img = cv2.imread('/content/gdrive/My Drive/images/girl.jpg')
print(img.shape)
gimg1 = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gimg1, cmap='gray')
```

```python
gMul = np.zeros((gimg1.shape[0], gimg1.shape[1]), dtype=np.ubyte)
```

```python
val = 3

for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    mulVal = np.int32(gimg1[h,w]) * val
    if mulVal > 255:
      gMul[h,w] = 255
    else:
      gMul[h,w] = mulVal

plt.imshow(gMul, cmap='gray')
```

```python
from google.colab.patches import cv2_imshow
from google.colab import drive
import numpy as np   # 배열 사용할 때 numpy
import cv2
import matplotlib.pyplot as plt
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/images/fcespnoise.bmp')
print(img.shape)
gimg = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
cv2_imshow(gimg)
```

```python
kernel = np.ones((3,3), np.uint8)
erosion = cv2.erode(gimg, kernel, iterations=1)
cv2_imshow(erosion)
```

```python
opening = cv2.dilate(erosion, kernel, iterations=1)
cv2_imshow(opening)
```

```python
dilate2 = cv2.dilate(opening, kernel, iterations=1)
cv2_imshow(dilate2)
```

```python
erode2 = cv2.erode(dilate2, kernel, iterations=1)
cv2_imshow(erode2)
```

```python
from google.colab.patches import cv2_imshow
from google.colab import drive
import numpy as np   # 배열 사용할 때 numpy
import cv2
import matplotlib.pyplot as plt
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/images/lena.jpg')
print(img.shape)
gimg = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
cv2_imshow(gimg)
```

```python
def saturation(value):
  if(value > 255):
    value = 255
  elif (value <0):
    value = 0
  return value
```

```python
kernel3 = np.ones((3,3),np.float32)/9
```

```python
output = cv2.filter2D(gimg,-1,kernel3)
cv2_imshow(output)
```

```python
gOut = np.zeros((gimg.shape[0], gimg.shape[1]), dtype = np.ubyte)
```

```python
for h in range(1, gimg.shape[0]-1):
  for w in range(1, gimg.shape[1]-1):
    sum = 0
    for i in range(-1,2):
      for j in range(-1,2):
        sum += gimg[h+i, w+j]*kernel3[i+1, j+1]
    sum = saturation(sum)
    gOut[h,w] = sum

cv2_imshow(gOut)
```
